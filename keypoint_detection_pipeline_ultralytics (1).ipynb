```python
model = YOLO("yolo11n-pose.pt")

video_path = "/content/drive/MyDrive/DAT255_Semesterproject/pipeline_test_videos/IMG_1208.MOV"

results = model.track(source=video_path, show=True, save=True, stream=True)

video_keypoints = np.ndarray((49, 17, 2))
for i, result in enumerate(results):
    normalized_keypoints = result.keypoints.xyn.cpu().numpy()
    video_keypoints[i] = normalized_keypoints
    result.show()  # display to screen
```

```python
print(video_keypoints.shape)
```

```python
for i, frame in enumerate(video_keypoints):
  print(frame)
```

```python
x = np.zeros((49, 34))
```

```python
for i, frame in enumerate(video_keypoints):
  for j, keypoint in enumerate(frame):
    x[i][j*2] = keypoint[0]
    x[i][j*2+1] = keypoint[1]
```
